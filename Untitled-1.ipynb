```python
import pandas as pd
import numpy as np
import yfinance as yf
import yahoo_fin.stock_info as si
import mysql.connector
from datetime import datetime, date
from random import randint
import glob
from decimal import Decimal, getcontext
getcontext().prec = 14
```

```python
def novaMovimentacao(ticker, data, valor, quantidade, tipo):
    val = [ticker, data, valor, quantidade, tipo]
    df = pd.DataFrame([val], columns = ['ticker', 'data', 'valor', 'quantidade', 'tipo'])
    arq = "D:\Projetos\PeDeMeia\movimentacao\\"+str(randint(0,1000))+str(datetime.now().strftime('%Y-%m-%d'))
    df.to_csv(arq, sep=";", index=False)
```

```python
def consultaMovimentacao():
    listaMov = []
    for arq in sorted(glob.glob("./movimentacao/*")):
        df = pd.read_csv(arq, index_col=None, header=0, sep=";")
        listaMov.append(df)
    
    dfMov = pd.concat(listaMov, axis=0, ignore_index=True)
    dfMov['data'] = pd.to_datetime(dfMov['data'], format="%Y-%m-%d")
    return dfMov
```

```python
novaMovimentacao('BABA','2021-10-15','451.00','0.54','V')
```

```python
dfMov = consultaMovimentacao()
dfMov.loc[dfMov['tipo'] == 'V', 'quantidade'] = dfMov['quantidade']*(-1)
dfMov = dfMov[dfMov['ticker'] == 'BABA']
```

```python
dfMovQTD = pd.pivot_table(dfMov, values='quantidade', index='data', columns=dfMov['ticker'].str.upper(), aggfunc=sum, fill_value=0)
dfMovQTD
```
